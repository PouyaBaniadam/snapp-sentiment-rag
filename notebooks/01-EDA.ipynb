```python
train['length'] = train['comment'].apply(lambda x: len(str(x).split()))
train['length'].describe()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "../assets/vazir.ttf"

happy_text = " ".join(train[train['label']=="HAPPY"]['comment'])
sad_text = " ".join(train[train['label']=="SAD"]['comment'])

happy_wc = WordCloud(
    width=1200,
    height=600,
    font_path=font_path,
    background_color="white"
).generate(happy_text)

sad_wc = WordCloud(
    width=1200,
    height=600,
    font_path=font_path,
    background_color="white"
).generate(sad_text)

plt.imshow(happy_wc)
plt.axis("off")
plt.show()

plt.imshow(sad_wc)
plt.axis("off")
plt.show()
```
